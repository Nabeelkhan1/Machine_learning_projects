```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
def convert_sqft_to_num(x):
    tokens=x.split('-')
    if len(tokens)==2:
        return(float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
def remove_outliers(df):
    df_out= pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        st=np.std(subdf.price_per_sqft)
        reduced_df = subdf[(subdf.price_per_sqft>(m-st))& (subdf.price_per_sqft<=(m+st))]
        df_out=pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out  
df6=remove_outliers(df5)
df6.shape
```

```python
def scatter_plot_chart(df,location):
    bhk2= df[(df.location==location)& (df.bhk==2)]
    bhk3=df[(df.location==location)& (df.bhk==3)]
    matplotlib.rcParams['figure.figsize']=(15,10)
    plt.scatter(bhk2.total_sqft,bhk2.price,color='blue',label='2bhk',s=50)
    plt.scatter(bhk3.total_sqft,bhk3.price,color='green',label='3bhk',s=50)
    plt.xlabel('total sqft area')
    plt.ylabel('price')
    plt.title('location')
    plt.legend()
    
scatter_plot_chart(df6,'Hebbal')
```

```python
def remove_bhk_outliers(df):
    exclude_indices=np.array([])
    for location,location_df in df.groupby('location'):
        bhk_stats={}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk]={
                'mean':np.mean(bhk_df.price_per_sqft),
                'std':np.std(bhk_df.price_per_sqft),
                'count':bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats=bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices=np.append(exclude_indices,bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')
        
df7=remove_bhk_outliers(df6)
df7.shape
```

# Model using pycaret library


```python
# Importing necessary libraries
from pycaret.regression import *

# Loading dataset
data=df11
data


```

```python

# Initializing the setup
regression_setup = setup(data, target='price', session_id=123)

```

```python
# Comparing different models
best_model = compare_models()

```

```python
# Creating a model
model = create_model('br')  # 'lr' stands for Linear Regression

```

```python
# Tuning the model
tuned_model = tune_model(model)

```

```python
# Plotting the model
plot_model(tuned_model, plot='residuals')

```

```python
# Finalizing the model
final_model = finalize_model(tuned_model)
```

```python

# Predicting on new data
predictions = predict_model(final_model, data=data)
```
